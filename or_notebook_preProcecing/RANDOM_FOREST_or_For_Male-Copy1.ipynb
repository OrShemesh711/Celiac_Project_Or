```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.preprocessing import Imputer
```

```python
#load it back:
df = pd.read_pickle('DATA_FREQ_VDJ_CDR3_98')
df.head()
```

```python
# Load the library with the iris dataset
from sklearn.datasets import load_iris

# Load scikit's random forest classifier library
from sklearn.ensemble import RandomForestClassifier

# Load pandas
import pandas as pd

# Load numpy
import numpy as np

# Set random seed
np.random.seed(0)
```

## Create Training And Test Data

```python
# Create a new column that for each row, generates a random number between 0 and 1, and
# if that value is less than or equal to .75, then sets the value of that cell as True
# and false otherwise. This is a quick and dirty way of randomly assigning some rows to
# be used as the training data and some as the test data.
df['is_train'] = np.random.uniform(0, 1, len(df)) <= .75
df = df.drop(['SUBJECTn','SEX'],axis=1)
# View the top 5 rows
df.head()
```

```python
# Create our imputer to replace missing values with the mean e.g.
imp = Imputer(missing_values='NaN', strategy='mean', axis=0)
imp = imp.fit(df.iloc[:,1:-1])
df.iloc[:,1:-1] = imp.transform(df.iloc[:,1:-1])


# Impute our data, then train
#X_train_imp = imp.transform(X_train)
```

```python
df
```

```python
# Create two new dataframes, one with the training rows, one with the test rows
train, test = df[df['is_train']==True], df[df['is_train']==False]
```

```python
# Show the number of observations for the test and training dataframes
print('Number of observations in the training data:', len(train))
print('Number of observations in the test data:',len(test))
```

## Procces data

```python
# Create a list of the feature column's names
features = df.columns[1:-1]

# View features
features
```

```python
# train['species'] contains the actual species names. Before we can use it,
# we need to convert each species name into a digit. So, in this case there
# are three species, which have been coded as 0, 1, or 2.
y = pd.factorize(train['CLIN'])[0]

# View target
y
```

```python
v = pd.factorize(train['CLIN'])
v
```

## TRAIN THE RANDOM FOREST CLASSIFIER

```python
# Create a random forest Classifier. By convention, clf means 'Classifier'
clf = RandomForestClassifier(n_jobs=2, random_state=0)

# Train the Classifier to take the training features and learn how they relate
# to the training y (the species)
clf.fit(train[features], y)
```

## Apply classifier to test data

```python
# Apply the Classifier we trained to the test data (which, remember, it has never seen before)
clf.predict(test[features])
```

```python
# View the predicted probabilities of the first 10 observations
clf.predict_proba(test[features])[0:10]
```

## evaluate classifier

## VIEW FEATURE IMPORTANCE

```python
# View a list of the features and their importance scores
feature_importance = list(zip(train[features], clf.feature_importances_))
```

## EVALUATING ERROR

```python

y_test =pd.factorize(test['CLIN'])[0]
display(y_test)
y_pred = clf.predict(test[features])
display(y_pred)
```

```python
from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))  
```
