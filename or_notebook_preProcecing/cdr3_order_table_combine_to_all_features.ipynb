## take the subject and sex coulmns (save as 'subject_sex_coulmns')

```python
# look on table for take the relevant coulmn name
cdr2_table = pd.read_csv('cdr_table.csv',sep='\t' ,quoting=csv.QUOTE_NONE)
cdr2_table.head()
```

```python
subject_sex = cdr2_table.loc[:,{'Unnamed: 0','Unnamed: 2'}]
subject_sex.rename({'Unnamed: 0':'SUBJECT' ,'Unnamed: 2':'SEX' }, inplace = True,axis='columns')
subject_sex = subject_sex.drop(subject_sex.index[[0,1]])
subject_sex.head()
```

```python
# save file
import csv
subject_sex.to_csv('subject_sex_coulmns.csv',sep='\t')
```
