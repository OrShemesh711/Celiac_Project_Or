# Get and combine feature table

Import new tables and packages 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Import table
import csv
Dfamily_DseqL = pd.read_table('Dfamily_DseqL_usage.tab',sep='\t')
R_FRAME = pd.read_table('FRAME_usage.tab',sep='\t')
Dfamily_FRAME = pd.read_table('Dfamily_FRAME_usage.tab',sep='\t')
Dgene_DseqL = pd.read_table('Dgene_DseqL_usage.tab',sep='\t')
```

Look on table before combining

```python
# look on the tables before combining
DF = Dgene_DseqL  #change to relevant table
display(DF.head())

# Check HOW MUCH SUBJECTS WE HAVE
UNIQ_SAMP = np.unique(DF.loc[:,'SAMPLE'])
print('number of subjects: ',len(UNIQ_SAMP))

# Check and drop nan values
print('Have nan values before dropping:' ,DF.loc[:,'SAMPLE'].isnull().values.any() )
# drop NA values- when we have no define gene
DF = DF.dropna(subset=['GENE'])
print('Have nan values after dropping:' ,DF.loc[:,'SAMPLE'].isnull().values.any() )
```

Create frequency column for each sample

```python


# Script for create frequency column for each sample
# check the unique variables
FN = DF  # relevant file name (FN)
UNIQ = np.unique(FN.loc[:,'GENE'])         ## change for the relevant coulmn name
display(UNIQ)
display(FN.head())
# create the first coulmn for the merging
name1 = UNIQ[0]
VFSUB1 = FN.loc[FN['GENE'] == name1 ]      ## change for the relevant coulmn name
coulmn_name1 = UNIQ[0] 
VFSUB1 = VFSUB1.assign(**{coulmn_name1 : VFSUB1.SEQ_FREQ})
VFSUB1= VFSUB1.loc[:,['SAMPLE',coulmn_name1]]
display(VFSUB1.head())
# create the other tables and combine them togther
len_uniq = len(UNIQ)
FN_all = VFSUB1
for i in range(1,len_uniq) :
    print(i)
    name = UNIQ[i]
    print(name)
    VFSUB2 = FN.loc[FN['GENE'] == name]     ## change for the relevant coulmn name
    coulmn_name = UNIQ[i]
    VFSUB2 = VFSUB2.assign(**{coulmn_name : VFSUB2.SEQ_FREQ})
    VFSUB2 = VFSUB2.loc[:,['SAMPLE',coulmn_name]]
    FN_all = pd.merge(FN_all , VFSUB2, how='outer',on=['SAMPLE'])
print('\n ~~~~~FN_all~~~~~')
display(FN_all)    

# sort by sample (alphabetic)
FN_SORT = FN_all.sort_values('SAMPLE')
print('\n ~~~~~sort by sample~~~~~')
display(FN_SORT)
```

Look columns name for match the 'on' category


```python
# Look columns name for match the 'on' category
merge_all = pd.read_pickle('DF_MERGE_TEMP')
# take name for subjects columns
subject_name_merge = merge_all.columns.values[0]
print('subject column name: ',subject_name_merge)
# change subject name in the new table
FN_SORT = FN_SORT.rename(columns={'SAMPLE': subject_name_merge})
```

Call to old merged table and add the new table

```python
# call for old merged table
merge_all = pd.read_pickle('DF_MERGE_TEMP')
# merging
merge_all = pd.merge(merge_all , FN_SORT, how='outer',on=[subject_name_merge])
# fill zeros in NA 
merge_all.iloc[:,3:] = merge_all.iloc[:,3:].fillna(0)
df_merge = merge_all
```

```python
# Check if we have nan values after enter zeros
print('We have nan values after zeros? ' ,df_merge.iloc[:,3:].isnull().values.any() )
```

SAVE FILE

```python
# save dataframe of merge all 
df_merge.to_pickle('DF_MERGE_TEMP')  # where to save it, usually as a .pkl
```
